## What is a Heuristic?

A **heuristic** is a problem-solving approach that uses practical methods to find good-enough solutions when exact methods are too slow or impractical. Unlike exact optimization software such as Pyomo or CPLEX, heuristics are not guaranteed to find the optimal solution, but they excel at finding high-quality solutions in reasonable time.

**Key insight**: Think of heuristics as "smart shortcuts" - they trade the guarantee of finding the absolute best solution for the ability to find a very good solution much faster.

### The advantages of heuristics are:
- **Easy to implement**: Often simpler to code than exact algorithms
- **Fast execution**: Can handle large problems that exact methods cannot solve in reasonable time
- **Flexible**: Can be adapted to various problem types
- **Scalable**: Performance degrades gracefully as problem size increases
- **Practical**: In most cases, the result is close enough to the optimal solution for real-world use

### Disadvantages to consider:
- No guarantee of finding the optimal solution
- Solution quality may vary between runs (for stochastic heuristics)
- Difficult to know how far the solution is from the optimum

## What is a Metaheuristic? And how is it different from a heuristic?

**Metaheuristics** are high-level problem-independent frameworks that guide other heuristics to explore the solution space more effectively. The key differences:

| Aspect | Heuristic | Metaheuristic |
|--------|-----------|---------------|
| **Scope** | Problem-specific | Problem-independent |
| **Flexibility** | Designed for one problem | Can be adapted to many problems |
| **Complexity** | Usually simpler | More sophisticated strategies |
| **Examples** | Greedy algorithm for TSP | Genetic Algorithms (work on any problem) |

**Important**: A metaheuristic is like a recipe template that can be customized for different dishes (problems), while a heuristic is a specific recipe for one dish.

## What types of metaheuristics are there?

Metaheuristics can be classified by their inspiration and approach:

### Nature-Inspired Metaheuristics:
- **Genetic Algorithms (GA)**: Inspired by biological evolution and natural selection
- **Ant Colony Optimization (ACO)**: Based on ant foraging behavior
- **Particle Swarm Optimization (PSO)**: Mimics bird flocking or fish schooling
- **Simulated Annealing (SA)**: Inspired by the metallurgical annealing process

### Search-Based Metaheuristics:
- **Tabu Search**: Uses memory structures to avoid revisiting solutions
- **Variable Neighborhood Search**: Systematically changes neighborhood structures

### Learning-Based:
- **Reinforcement Learning**: Learns optimal strategies through trial and error

# Algoritmos Genéticos

## Introducción

Los algoritmos genéticos (AG) son una clase de algoritmos de optimización y búsqueda basados en los principios de la genética y la selección natural de Darwin. Forman parte de los algoritmos evolutivos y se utilizan para resolver problemas complejos donde los métodos tradicionales pueden ser ineficientes o inadecuados. Su capacidad para explorar grandes espacios de búsqueda y encontrar soluciones óptimas o cercanas al óptimo los hace valiosos en diversas aplicaciones.

## ¿Qué es un algoritmo genético?

Un algoritmo genético es un método de optimización inspirado en el proceso de evolución biológica. Emula la forma en que las poblaciones de organismos evolucionan a través de generaciones mediante mecanismos de selección natural, cruce y mutación. El objetivo es encontrar la mejor o las mejores soluciones a un problema específico, representado mediante una función de aptitud que evalúa la calidad de cada solución candidata.

## ¿Cómo funciona un algoritmo genético?

El funcionamiento de un algoritmo genético se basa en los siguientes pasos fundamentales:

1. **Inicialización de la población**: Se crea una población inicial de soluciones candidatas (individuos) de forma aleatoria o utilizando conocimientos previos.

2. **Evaluación**: Cada individuo de la población es evaluado mediante la función de aptitud, que asigna un valor numérico representando su calidad o adaptación al problema.

3. **Selección**: Se eligen los individuos más aptos para reproducirse. Los métodos de selección pueden ser proporcionales a la aptitud, torneos, entre otros.

4. **Cruce (recombinación)**: Los individuos seleccionados (padres) se combinan para producir nuevos individuos (descendientes), intercambiando partes de su información genética. Esto permite explorar nuevas áreas del espacio de búsqueda.

5. **Mutación**: Se aplican cambios aleatorios a los descendientes para mantener la diversidad genética y evitar la convergencia prematura hacia soluciones subóptimas.

6. **Reemplazo**: Los nuevos individuos sustituyen parcial o totalmente a la población actual, formando una nueva generación.

7. **Iteración**: Se repiten los pasos de evaluación, selección, cruce y mutación durante varias generaciones hasta cumplir con un criterio de parada, como alcanzar un número máximo de generaciones o una aptitud satisfactoria.

## Componentes clave de un algoritmo genético

- **Población**: Conjunto de individuos que representan posibles soluciones al problema. El tamaño de la población afecta la diversidad y la convergencia del algoritmo.

- **Representación genética**: Forma en que se codifican las soluciones, comúnmente como cadenas de bits, números reales, permutaciones, entre otros, dependiendo del problema.

- **Función de aptitud**: Mide la calidad de cada individuo en relación con el objetivo del problema. Es esencial para guiar el proceso evolutivo hacia mejores soluciones.

- **Operadores genéticos**:
  - **Selección**: Mecanismo para elegir los individuos que participarán en la reproducción. Favorece a los más aptos, pero permite que individuos menos aptos también tengan oportunidad para mantener la diversidad.
  - **Cruce**: Combina información genética de dos o más padres para producir descendientes. Métodos comunes incluyen cruce de un punto, de dos puntos y uniforme.
  - **Mutación**: Introduce variaciones aleatorias en los individuos. Ayuda a explorar nuevas soluciones y a evitar la estancamiento en óptimos locales.

- **Parámetros del algoritmo**: Incluyen la tasa de cruce, tasa de mutación, tamaño de la población y criterios de parada, que deben ser ajustados cuidadosamente para el problema en cuestión.

## ¿Cómo aprende un algoritmo genético?

El algoritmo genético aprende y mejora las soluciones a través de un proceso iterativo que simula la evolución natural:

- **Selección de los más aptos**: Al favorecer la reproducción de individuos con mayor aptitud, el algoritmo incrementa la probabilidad de que las características beneficiosas se transmitan a las siguientes generaciones.

- **Recombinación de información**: El cruce permite combinar las mejores características de diferentes individuos, explorando nuevas soluciones que podrían ser superiores a las anteriores.

- **Introducción de variabilidad**: La mutación añade diversidad genética, permitiendo al algoritmo escapar de óptimos locales y explorar regiones inexploradas del espacio de búsqueda.

- **Adaptación progresiva**: A lo largo de las generaciones, la población se adapta al problema, mejorando gradualmente la calidad de las soluciones hasta alcanzar o aproximarse al óptimo global.

## Aplicaciones de los algoritmos genéticos

Los algoritmos genéticos se han aplicado exitosamente en diversos campos:

## Understanding the GA Workflow Through Pseudocode

Below, we present the genetic algorithm in pseudocode divided into two parts: **initialization** and the **main evolutionary loop**. Study each section carefully and think about what each step accomplishes.

### Part 1: Initialization Phase

```
BEGIN GeneticAlgorithm
    // Define algorithm parameters
    PopulationSize ← N              // How many individuals in each generation?
    MaxGenerations ← G              // When should we stop evolving?
    CrossoverRate ← Pc              // Probability of breeding (typically 0.7-0.95)
    MutationRate ← Pm               // Probability of random changes (typically 0.01-0.1)
    ElitismCount ← E                // How many top individuals to preserve?

    // Initialize population
    Population ← InitializePopulation(PopulationSize)
    
    // Evaluate initial population
    FOR each individual IN Population DO
        individual.Fitness ← EvaluateFitness(individual)
    END FOR

    Generation ← 0
```

**Key Questions to Consider**:
1. **PopulationSize**: What happens if this is too small? Too large?
2. **CrossoverRate**: Why not always do crossover (why not 100%)?
3. **MutationRate**: Why is this typically much smaller than crossover rate?
4. **ElitismCount**: What problem does elitism solve?

**Your Task**: Think about how you would initialize a population for the maze problem. What should each "individual" look like?

### Part 2: Main Evolutionary Loop

This is where the "evolution" happens. Each iteration represents one generation.

```
    WHILE Generation < MaxGenerations AND NOT TerminationConditionMet DO
        // Step 1: Selection - Choose who gets to reproduce
        MatingPool ← SelectIndividuals(Population)
        
        // Step 2: Crossover (Recombination) - Create offspring
        OffspringPopulation ← []
        FOR i FROM 0 TO (PopulationSize - ElitismCount) STEP 2 DO
            Parent1 ← MatingPool.Select()
            Parent2 ← MatingPool.Select()
            IF Random() < CrossoverRate THEN
                Child1, Child2 ← Crossover(Parent1, Parent2)
            ELSE
                Child1 ← Parent1.Clone()
                Child2 ← Parent2.Clone()
            END IF
            OffspringPopulation.Add(Child1)
            OffspringPopulation.Add(Child2)
        END FOR
        
        // Step 3: Mutation - Introduce random variations
        FOR each individual IN OffspringPopulation DO
            IF Random() < MutationRate THEN
                Mutate(individual)
            END IF
        END FOR
        
        // Step 4: Evaluate Offspring - Calculate fitness
        FOR each individual IN OffspringPopulation DO
            individual.Fitness ← EvaluateFitness(individual)
        END FOR
        
        // Step 5: Replacement - Create new generation
        Population ← ApplyElitism(Population, OffspringPopulation, ElitismCount)
        
        // Increment generation counter
        Generation ← Generation + 1
    END WHILE

    // Return the best solution found
    BestIndividual ← FindBestIndividual(Population)
    RETURN BestIndividual
END GeneticAlgorithm
```

**Understanding Each Step**:

1. **Selection**: Not all individuals get to reproduce - fitter ones should have higher chances
   - *Question*: Should the worst individuals ever get selected? Why or why not?

2. **Crossover**: Combines genetic material from two parents
   - *Question*: If Parent1 has a good first half and Parent2 has a good second half, how might crossover help?

3. **Mutation**: Makes small random changes
   - *Question*: Why is mutation important even if we have crossover?

4. **Evaluation**: Every new individual needs a fitness score
   - *Question*: What makes a good fitness function for the maze problem?

5. **Replacement**: Decide which individuals survive to the next generation
   - *Question*: Why might we want to keep the best individuals from the previous generation (elitism)?

## Applying GAs to the Maze Problem

### Understanding the Problem Domain

Before diving into implementation, let's think about how to represent maze navigation as a genetic algorithm problem.

### Genome Representation

**The Challenge**: How do we encode a path through a maze as a "genome" that can evolve?

**Option 1 - Direction Sequence**: 
- Genome = list of moves like `['UP', 'RIGHT', 'RIGHT', 'DOWN', ...]`
- *Pros*: Simple, easy to understand
- *Cons*: May try invalid moves (walking into walls)

**Option 2 - Numeric Encoding**:
- Genome = list of numbers like `[0, 1, 1, 2, ...]` where 0=UP, 1=RIGHT, 2=DOWN, 3=LEFT
- *Pros*: Easy to mutate and crossover
- *Cons*: Still may generate invalid paths

**Option 3 - Valid Move Indices**:
- At each position, only store indices of valid moves
- *Pros*: Always generates valid moves
- *Cons*: More complex to implement

**Question**: Which representation would you choose and why? Consider the tradeoffs.

### Fitness Function Design

The fitness function is crucial - it guides evolution toward better solutions. For the maze problem, consider:

**What should we reward?**
- Getting closer to the exit
- Reaching the exit (big bonus!)
- Exploring new areas

**What should we penalize?**
- Hitting walls
- Going in circles (revisiting positions)
- Taking too many steps
- Staying far from the goal

**Design Challenge**: How would you balance these factors? If you make the wall penalty too harsh, the algorithm might not explore enough. Too lenient, and it wastes time on invalid paths.

**Hint**: Think about using a combination of:
1. **Distance-based reward**: How close did we get to the exit?
2. **Exploration bonus**: How many unique cells did we visit?
3. **Efficiency penalty**: How many steps did we take?
4. **Success bonus**: Did we reach the exit?

### Visualization

Below you'll see a maze generated using a depth-first search algorithm. The entrance is at the top-left, and the exit is at the bottom-right.

**Study the maze structure**: 
- White cells are paths
- Black cells are walls
- Notice how paths wind around - there's no straight line to the exit

````python
import matplotlib.pyplot as plt
import numpy as np
import random
from matplotlib.colors import ListedColormap

def create_maze(dim):
    # Create a grid filled with walls (1)
    # Set the seed to ensure reproducibility
    np.random.seed(2)
    maze = np.ones((dim * 2 + 1, dim * 2 + 1), dtype=int)

    # Define the starting point
    x, y = (0, 0)
    maze[2 * x + 1, 2 * y + 1] = 0  # Mark start as open path

    # Initialize the stack with the starting point for DFS
    stack = [(x, y)]
    
    while stack:
        x, y = stack[-1]

        # Define possible directions (right, down, left, up)
        directions = [(0, 1), (1, 0), (0, -1), (-1, 0)]
        random.shuffle(directions)  # Randomize order for more organic mazes

        for dx, dy in directions:
            nx, ny = x + dx, y + dy
            # Check if next cell is within bounds and is a wall
            if 0 <= nx < dim and 0 <= ny < dim and maze[2 * nx + 1, 2 * ny + 1] == 1:
                # Break wall to create path
                maze[2 * nx + 1, 2 * ny + 1] = 0
                maze[2 * x + 1 + dx, 2 * y + 1 + dy] = 0
                stack.append((nx, ny))  # Move to the next cell
                break
        else:
            stack.pop()  # Backtrack if no unvisited neighbors

    # Create entrance and exit points
    maze[1, 0] = 0  # Entrance
    maze[-2, -1] = 0  # Exit

    return maze

def display_maze(maze):
    cmap = ListedColormap(['white', 'black', 'green', 'black'])
    plt.figure(figsize=(6, 6))
    plt.pcolor(maze[::-1], cmap=cmap, edgecolors='k', linewidths=2)
    plt.gca().set_aspect('equal')
    plt.xticks([])
    plt.yticks([])
    plt.title('Maze with Entrance and Exit')
    plt.show()

# Define the dimension of the maze (adjustable)
dim = 25  # This can be any positive integer
maze = create_maze(dim)
display_maze(maze)
````

### 3. `select(population, fitnesses)`

**Purpose**: Choose individuals from the population to become parents for the next generation.

- Probability of selection proportional to fitness

**Rank-Based Selection**:
- Sort by fitness
- Selection probability based on rank, not absolute fitness

**Think about**:

- Should you allow the same individual to be selected multiple times?
- How does selection pressure affect convergence speed?

**Common pitfalls**:
- ❌ Only selecting the top N individuals (too much selection pressure)

- ❌ Returning fewer individuals than needed for creating next generation
- ✅ Maintain diversity while favoring fitter individuals

### 4. `crossover(parent1, parent2)`

**Purpose**: Combine genetic material from two parents to create offspring.

**What it should do**:
- Take two parent genomes as input
- Return one or two child genomes
- Children should inherit traits from both parents

**Crossover methods to consider**:

**Single-Point Crossover**:
```
Parent1: [A, B, C, D, E, F]
Parent2: [1, 2, 3, 4, 5, 6]
         Cut point ↓
Child1:  [A, B, C, 4, 5, 6]
Child2:  [1, 2, 3, D, E, F]
```

**Two-Point Crossover**:
```
Parent1: [A, B, C, D, E, F]
Parent2: [1, 2, 3, 4, 5, 6]
         ↓ Cut    Cut ↓
Child1:  [A, 2, 3, 4, E, F]
Child2:  [1, B, C, D, 5, 6]
```

**Uniform Crossover**:
- Each gene has 50% chance of coming from either parent

**Think about**:
- Does the crossover point matter for the maze problem?
- Should you return one child or two?
- What if parents have different lengths? (shouldn't happen if implemented correctly)

**Common pitfalls**:
- ❌ Always using the same crossover point
- ❌ Creating children that are identical to one parent
- ❌ Not maintaining genome length
- ✅ Randomize crossover points to explore different combinations

**Testing checklist**:
- [ ] Returns valid genome(s)
- [ ] Children have genes from both parents
- [ ] Children have the same length as parents

### 5. `mutate(individual, mutation_rate)`

**Purpose**: Introduce random changes to maintain genetic diversity and explore new solutions.

**What it should do**:
- Iterate through each gene in the genome
- With probability `mutation_rate`, change that gene to a random valid value
- Modify the individual in-place or return a new mutated individual

**Mutation strategies**:

**Random Resetting**:
- Replace gene with a completely random new value

**Swap Mutation**:
- Swap two random positions in the genome

**Scramble Mutation**:
- Randomly shuffle a subset of genes

**Think about**:
- What is an appropriate mutation rate? (typically 0.01 to 0.1)
- Should mutation be completely random or biased in some direction?
- How does mutation rate affect exploration vs exploitation?

**Common pitfalls**:
- ❌ Mutation rate too high (destroys good solutions)
- ❌ Mutation rate too low (population gets stuck)
- ❌ Mutations that create invalid gene values
- ✅ Start with low mutation rate (1-5%) and adjust based on results

**Testing checklist**:

**Common pitfalls**:

## Implementation Strategy

### Recommended Development Order:

1. **Start with `generate_population`** - This is the simplest function and gives you data to work with
2. **Implement `reward` next** - You need this to test if your fitness function makes sense
3. **Test genome representation** - Create a few hand-crafted genomes and verify the reward function works as expected
4. **Implement `mutate`** - Easy to test independently
5. **Implement `crossover`** - Also testable independently
6. **Implement `select`** - Requires fitness values from reward function
7. **Finally, implement `evolve`** - Brings everything together

### Debugging Tips:

**If the algorithm isn't improving**:
- Check if fitness function actually rewards better solutions
- Verify selection is favoring fitter individuals
- Try increasing mutation rate temporarily
- Print best fitness each generation to see if there's any progress

**If the algorithm converges too quickly to a suboptimal solution**:
- Increase population size
- Increase mutation rate
- Decrease selection pressure
- Add more diversity in initialization

**If the algorithm is too slow**:
- Reduce population size
- Reduce genome length (if possible)
- Optimize fitness function (it's called most frequently)
- Consider vectorizing operations with NumPy
